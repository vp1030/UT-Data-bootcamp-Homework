# Purchasing Analysis (Gender)

```python
#Purchasing Analysis (Gender)

#The below each broken by gender
#group by gender
gender_groups = pymoli_pd.groupby(["Gender"])

#Purchase Count
purchase_count = gender_groups["SN"].count()
#print(purchase_count)

#Average Purchase Price
avgpurchase_price = gender_groups["Price"].mean()
#print(avgpurchase_price)

#Total Purchase Value
purchase_value = gender_groups["Price"].sum()
#print(purchase_value)

#Normalized Totals
duplicate_data = pymoli_pd.drop_duplicates(subset='SN', keep="first")
grouped_duplicate_data = duplicate_data.groupby(["Gender"])

Norm = (gender_groups["Price"].sum() / grouped_duplicate_data["SN"].count())
```

```python
# Create a new table consolodating above calculations
purchase_analysis = pd.DataFrame({"Purchase Count": purchase_count,
                                   "Average Purchase Price": avgpurchase_price,
                                   "Total Purchase Value": purchase_value,
                                   "Normalized Totals": Norm})
purchase_analysis = purchase_analysis[["Purchase Count", 
                                         "Average Purchase Price", 
                                         "Total Purchase Value", 
                                         "Normalized Totals"]]

# Improve formatting before outputting
purchase_analysis["Average Purchase Price"] = purchase_analysis["Average Purchase Price"].map("${0:,.2f}".format)
purchase_analysis["Total Purchase Value"] = purchase_analysis["Total Purchase Value"].map("${0:,.2f}".format)
purchase_analysis["Normalized Totals"] = purchase_analysis["Normalized Totals"].map("${0:,.2f}".format)

purchase_analysis
```

# Purchasing Analysis (Age)

```python
# Create a new table consolodating calculations below with the age ranges
#Copy the bins from above
bins = [0, 10, 15, 20, 25, 30, 35, 40, 100]
labels = ['Under 10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

#Make bin a datafram 
bin_df = pymoli_pd.copy()
bin_df["Age Groups"] = pd.cut(bin_df["Age"], bins, labels=labels)
bin_1 = pd.cut(bin_df["Age"], bins, labels=labels)
grouped_bin = bin_df.groupby(["Age Groups"])

#Purchase Count
agepurchase_count = grouped_bin["Age"].count()
#print(agepurchase_count)

#Average Purchase Price
age_avgpurchase_price = grouped_bin["Price"].mean()
#print(age_avgpurchase_price)

#Total Purchase Value
age_purchase_value = grouped_bin["Price"].sum()
#print(age_purchase_value)

#Normalized Totals
#de-dupe 
duplicate = pymoli_pd.drop_duplicates(subset = 'SN', keep = "first")
duplicate["Age Groups"] = pd.cut(duplicate["Age"], bins, labels=labels)
duplicate = duplicate.groupby(["Age Groups"])

norm_totals = (grouped_bin["Price"].sum()/duplicate["SN"].count())
norm_totals

#New dataframe
age_analysis = pd.DataFrame({"Purchase Count": agepurchase_count,
                            "Average Purchase Price": age_avgpurchase_price,
                            "Total Purchase Value": age_purchase_value,
                            "Normalized Totals": norm_totals})
age_analysis = age_analysis[["Purchase Count", 
                                         "Average Purchase Price", 
                                         "Total Purchase Value", 
                                         "Normalized Totals"]]

# Improve formatting before outputting 
age_analysis["Average Purchase Price"] = age_analysis["Average Purchase Price"].map("${0:,.2f}".format)
age_analysis["Total Purchase Value"] = age_analysis["Total Purchase Value"].map("${0:,.2f}".format)
age_analysis["Normalized Totals"] = age_analysis["Normalized Totals"].map("${0:,.2f}".format)

age_analysis
```

# Top Spenders

```python
#Top Spenders

#Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
#Group by SN
sn_group = pymoli_pd.groupby(["SN"])

#Purchase Count
sn_group_count = sn_group["Item ID"].count()
#sn_group_count

#top_spenders = pymoli_df.groupby(["SN"])
#top_spenders.count()

#Average Purchase Price
sn_group_avg = sn_group["Price"].mean()
#sn_group_avg
#top_spenders["Price"].mean()

#Total Purchase Value
sn_group_total = sn_group["Price"].sum()
#sn_group_total
#top_spenders["Price"].sum()

#Create new dateframe to output data
top_spenders = pd.DataFrame({"Purchase Count": sn_group_count, "Average Purchase Price": sn_group_avg,"Total Purchase Value":sn_group_total})

top_spenders = top_spenders[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]

top_spenders = top_spenders.sort_values("Total Purchase Value", ascending = False)


# Improve formatting before outputting 
top_spenders["Average Purchase Price"] = top_spenders["Average Purchase Price"].map("${0:,.2f}".format)
top_spenders["Total Purchase Value"] = top_spenders["Total Purchase Value"].map("${0:,.2f}".format)

top_spenders.head()
```

# Most Popular Items

```python
#Most Popular Items

#Identify the 5 most popular items by purchase count, then list (in a table):
item_group = pymoli_pd.groupby(["Item ID", "Item Name"])

#Purchase Count
item_group_count = item_group["Item ID"].count()
#item_group_count

#Item Price
item_group_price = item_group["Price"].mean()
#item_group_price
#top_spenders["Price"].mean()

#Total Purchase Value
item_group_total = item_group["Price"].sum()

#Create new dateframe to output data
top_items = pd.DataFrame({"Purchase Count": item_group_count, "Item Price": item_group_price,"Total Purchase Value":item_group_total})

top_items = top_items[["Purchase Count", "Item Price", "Total Purchase Value"]]

top_items = top_items.sort_values("Purchase Count", ascending = False)

# Improve formatting before outputting 
top_items["Item Price"] = top_items["Item Price"].map("${0:,.2f}".format)
top_items["Total Purchase Value"] = top_items["Total Purchase Value"].map("${0:,.2f}".format)


top_items.head()
```
